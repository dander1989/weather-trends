```python
import requests
import pandas as pd
```

```python
# Testing the API waters
r = requests.get('https://archive-api.open-meteo.com/v1/archive?latitude=39.0428&longitude=-89.9518&start_date=2000-01-01&end_date=2026-01-01&daily=temperature_2m_max,temperature_2m_min,temperature_2m_mean,precipitation_sum,snowfall_sum,wind_speed_10m_max&timezone=America%2FChicago&temperature_unit=fahrenheit&wind_speed_unit=mph&precipitation_unit=inch')
data = r.json()
print(data)
```

```python
bh_daily = data['daily']
```

```python
# Accessing all records of max temp
all_max_temps = (data['daily']['temperature_2m_max'])
all_max_temps
```

```python
# Accessing first record of max temp
day1_temp = (data['daily']['temperature_2m_max'][0])
day1_temp
```

```python
# Creating a dataframe from the daily temps
df = pd.DataFrame(bh_daily)
df
```

```python
# Identifying data types of DF
df.dtypes
```

```python
# Converting 'string' dates into datetime
df['time'] = pd.to_datetime(df['time'], format="%Y-%m-%d")
df = df.rename(columns= {'time':'date'})
df.dtypes
```

```python
# Splitting the date column into 'year', 'month', 'day' columns
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df
```

```python
# Create 'season' column and apply each month to a season.
season_dict = {1:"Winter", 2:"Winter", 3:"Spring", 4:"Spring", 5:"Spring", 6:"Summer", 7:"Summer", 8:"Summer", 9:"Fall", 10:"Fall", 11:"Fall", 12:"Winter"}
df['season'] = df['month'].map(season_dict)
df.head()
df[df['month'] == 7].head()
```
